#1. Análisis de facturación



```python
# 1. Análisis de facturación
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

print("Ingresos totales por tienda:")
print(ingresos_totales)

ingresos_totales.to_csv('ingresos_totales.csv', index=False)
```

# 2. Ventas por categoría

```python
# 2. Ventas por categoría
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()

ventas_categoria_tienda1.columns = ['Categoría', 'Ingreso Tienda 1']
ventas_categoria_tienda2.columns = ['Categoría', 'Ingreso Tienda 2']
ventas_categoria_tienda3.columns = ['Categoría', 'Ingreso Tienda 3']
ventas_categoria_tienda4.columns = ['Categoría', 'Ingreso Tienda 4']

ventas_categorias = ventas_categoria_tienda1.merge(
    ventas_categoria_tienda2, on='Categoría', how='outer'
).merge(
    ventas_categoria_tienda3, on='Categoría', how='outer'
).merge(
    ventas_categoria_tienda4, on='Categoría', how='outer'
).fillna(0)

print("Ventas por categoría para cada tienda:")
print(ventas_categorias)

ventas_categorias.to_csv('ventas_categorias.csv', index=False)
```
