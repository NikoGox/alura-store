#1. Análisis de facturación



```python
# 1. Análisis de facturación
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

print("Ingresos totales por tienda:")
print(ingresos_totales)

ingresos_totales.to_csv('ingresos_totales.csv', index=False)
```

Los ingresos totales de cada tienda muestran que Tienda 1 tiene el mayor ingreso con 1.150 millones, seguida de Tienda 2 con 1.116 millones. Las Tiendas 3 y 4 presentan ingresos similares, alrededor de 1.098 millones y 1.038 millones, respectivamente. Esto indica que, en términos generales, la Tienda 1 tiene un rendimiento más alto, aunque la diferencia no es drástica, lo que podría significar que todas las tiendas están manteniendo un nivel de ventas competitivo.

# 2. Ventas por categoría

```python
# 2. Ventas por categoría
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()

ventas_categoria_tienda1.columns = ['Categoría', 'Ingreso Tienda 1']
ventas_categoria_tienda2.columns = ['Categoría', 'Ingreso Tienda 2']
ventas_categoria_tienda3.columns = ['Categoría', 'Ingreso Tienda 3']
ventas_categoria_tienda4.columns = ['Categoría', 'Ingreso Tienda 4']

ventas_categorias = ventas_categoria_tienda1.merge(
    ventas_categoria_tienda2, on='Categoría', how='outer'
).merge(
    ventas_categoria_tienda3, on='Categoría', how='outer'
).merge(
    ventas_categoria_tienda4, on='Categoría', how='outer'
).fillna(0)

print("Ventas por categoría para cada tienda:")
print(ventas_categorias)

ventas_categorias.to_csv('ventas_categorias.csv', index=False)
```

El análisis de ventas por categoría revela que Electrodomésticos y Electrónicos son las categorías más fuertes en términos de ingresos, especialmente para Tienda 1 y Tienda 2, con ventas significativas en ambas. Sin embargo, Instrumentos musicales y Juguetes tienen ingresos más bajos, sugiriendo que estas categorías podrían necesitar más promoción o una oferta más atractiva para incrementar su rendimiento. Las diferencias en ingresos por tienda en categorías específicas también pueden reflejar variaciones en el mercado y las preferencias regionales.

# 3. Calificación promedio de la tienda


```python
# 3. Calificación promedio de la tienda
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
})

print("Calificación promedio por tienda:")
print(calificaciones_promedio)
```

Las calificaciones promedio de las tiendas están bastante equilibradas, con Tienda 3 teniendo la calificación más alta (4.05), seguida por Tienda 2 (4.04). Esto sugiere que los clientes están generalmente satisfechos con las compras y experiencias en estas tiendas. Sin embargo, Tienda 1 y Tienda 4 tienen calificaciones ligeramente más bajas, lo que podría implicar áreas de mejora en la atención al cliente o en la experiencia general de compra.

# 4. Productos más y menos vendidos

```python
# 4. Productos más y menos vendidos
ventas_productos_tienda1 = tienda['Producto'].value_counts().reset_index()
ventas_productos_tienda2 = tienda2['Producto'].value_counts().reset_index()
ventas_productos_tienda3 = tienda3['Producto'].value_counts().reset_index()
ventas_productos_tienda4 = tienda4['Producto'].value_counts().reset_index()

ventas_productos_tienda1.columns = ['Producto', 'Ventas Tienda 1']
ventas_productos_tienda2.columns = ['Producto', 'Ventas Tienda 2']
ventas_productos_tienda3.columns = ['Producto', 'Ventas Tienda 3']
ventas_productos_tienda4.columns = ['Producto', 'Ventas Tienda 4']

mas_vendidos_tienda1 = ventas_productos_tienda1.head(1)
menos_vendidos_tienda1 = ventas_productos_tienda1.tail(1)
mas_vendidos_tienda2 = ventas_productos_tienda2.head(1)
menos_vendidos_tienda2 = ventas_productos_tienda2.tail(1)
mas_vendidos_tienda3 = ventas_productos_tienda3.head(1)
menos_vendidos_tienda3 = ventas_productos_tienda3.tail(1)
mas_vendidos_tienda4 = ventas_productos_tienda4.head(1)
menos_vendidos_tienda4 = ventas_productos_tienda4.tail(1)

productos_extremos = pd.concat([
    mas_vendidos_tienda1.assign(Tienda='Tienda 1', Tipo='Más Vendido'),
    menos_vendidos_tienda1.assign(Tienda='Tienda 1', Tipo='Menos Vendido'),
    mas_vendidos_tienda2.assign(Tienda='Tienda 2', Tipo='Más Vendido'),
    menos_vendidos_tienda2.assign(Tienda='Tienda 2', Tipo='Menos Vendido'),
    mas_vendidos_tienda3.assign(Tienda='Tienda 3', Tipo='Más Vendido'),
    menos_vendidos_tienda3.assign(Tienda='Tienda 3', Tipo='Menos Vendido'),
    mas_vendidos_tienda4.assign(Tienda='Tienda 4', Tipo='Más Vendido'),
    menos_vendidos_tienda4.assign(Tienda='Tienda 4', Tipo='Menos Vendido')
])

print("Productos más y menos vendidos por tienda:")
print(productos_extremos)
```

El análisis de los productos más y menos vendidos destaca algunos productos clave que impulsan las ventas. Tienda 1 tiene el Microondas como su producto más vendido, lo que refleja una fuerte demanda de electrodomésticos, mientras que el Celular ABXY es el menos vendido. Tienda 2, Tienda 3 y Tienda 4 presentan productos con un rendimiento más variado, pero no hay un patrón claro debido a la falta de datos en algunos productos. Este análisis sugiere que las tiendas pueden beneficiarse de una mayor estrategia de promoción o ajuste en los productos menos vendidos, como el Juego de mesa y el Bloques de construcción.




# 5. Envío promedio por tienda

```python
# Promedio de costos de envío por tienda
envio_tienda1 = tienda['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

# Mostrar resultados
print("Promedio de costo de envío por tienda:")
print(f"Tienda 1: {envio_tienda1}")
print(f"Tienda 2: {envio_tienda2}")
print(f"Tienda 3: {envio_tienda3}")
print(f"Tienda 4: {envio_tienda4}")
```

En términos de costos de envío, Tienda 1 tiene el costo promedio más alto (26,018), seguido de cerca por Tienda 2 (25,216). Esto podría implicar que algunas tiendas están incurriendo en mayores costos logísticos que otras. Las tiendas deberían explorar oportunidades para optimizar sus procesos de envío y reducir estos costos, lo cual podría mejorar su rentabilidad. Tienda 3 y Tienda 4 tienen costos de envío más bajos, lo que sugiere que están manejando sus operaciones logísticas de manera más eficiente.
